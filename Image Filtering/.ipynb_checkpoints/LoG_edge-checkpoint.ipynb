```python
import numpy as np
import cv2
import matplotlib.pyplot as plt
import skimage
from skimage import data, io, filters
import math

# --- Part 2.2: Q. 1 ---

# load the image as grayscale
woman_img = "./woman.png"
img = cv2.imread(woman_img, cv2.IMREAD_GRAYSCALE)

# apply 5x5 Laplacian filter to the intensity channel
lapG = cv2.Laplacian(img, cv2.CV_32F, ksize=5)

# zero crossing (more precise than peaks of 1st derivative)
def LaplacianZC(img):
    ht, wt = img.shape
    zeros = np.zeros((ht,wt))
    for i in range(ht):
        for j in range(wt):
            if(img[i][j] == 0):
                zeros[i][j] = 255
            elif(img[i][j] < 0):
                zeros[i][j] = 255
    return zeros

zero_crossing = LaplacianZC(lapG)

# display images
plt.figure(figsize=(20, 20))
plt.subplot(131), plt.imshow(img, cmap="gray")
plt.title("Original Image"), plt.xticks([]), plt.yticks([])

plt.subplot(132), plt.imshow(lapG, cmap="gray")
plt.title("Image with Laplacian of Gaussian"), plt.xticks([]), plt.yticks([])

plt.subplot(133), plt.imshow(zero_crossing, cmap="gray")
plt.title("Zero-crossings of the Laplacian"), plt.xticks([]), plt.yticks([])
plt.show()

# --- END of Part 2.2: Q. 1 ---
```
