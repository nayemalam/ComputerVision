## Part 1.1
    Q. 1,2,3

```python
import numpy as np
import cv2
import matplotlib.pyplot as plt
import skimage
from skimage import data, io, filters
import math

# --- Part 1.1: Q. 1,2,3 ---
pepper_img = "./peppers.png"
npepper_img = "./gaussian_noise_peppers.png"

# image filtering
img = cv2.imread(pepper_img)
img2 = cv2.imread(npepper_img)

# filter image using 7x7 Gaussian
img2_gaussian = cv2.GaussianBlur(img2, (7,7), math.sqrt(2)) #  sqrt(2) is standard dev., 2 is variance

# box filter image using 7x7 
img2_box = cv2.blur(img2, (7,7))

# convert from BGR to RGB; so it's not blue colored pic
img2_true = cv2.cvtColor(img2, cv2.COLOR_BGR2RGB)
res_gaussian = cv2.cvtColor(img2_gaussian, cv2.COLOR_BGR2RGB)
res_box = cv2.cvtColor(img2_box, cv2.COLOR_BGR2RGB)

# SNR using skimage
snr1 = skimage.measure.compare_psnr(img, img2_gaussian,None)
snr2 = skimage.measure.compare_psnr(img, img2_box, None)

# plot
plt.figure(figsize=(20, 20))
plt.subplot(131), plt.imshow(img2_true)
plt.title("Clear Image"), plt.xticks([]), plt.yticks([])

plt.subplot(132), plt.imshow(res_gaussian)
plt.title("7x7 Gaussian"), plt.xticks([]), plt.yticks([])

plt.subplot(133), plt.imshow(res_box)
plt.title("7x7 Box"), plt.xticks([]), plt.yticks([])

plt.show()

print("Original vs. Gaussian SNR :", snr1)
print("Original vs. Box SNR      :", snr2)

# --- END of Part 1.1: Q. 1,2,3 ---
```

### Section 1, 1.1 - Q3 Answer:
Gaussian blur seems to give the superior result; because a higher SNR tends to be considerably better. <br>
Signal/Noise, higher would mean high signal and low noise.

## Part 1.1

```python
# --- Part 1.1: Q. 4,5,6 ---
salt_pepper = "./saltandpepper_noise_peppers.png"
img3 = cv2.imread(salt_pepper)

# filter image using 7x7 Gaussian
img3_gaussian = cv2.GaussianBlur(img3, (7,7), math.sqrt(2)) 

# median filter image using 7x7 
img3_median = cv2.medianBlur(img3, 7)

# convert from BGR to RGB; so it's not blue colored pic
img3_true = cv2.cvtColor(img3, cv2.COLOR_BGR2RGB)
res1_gaussian = cv2.cvtColor(img3_gaussian, cv2.COLOR_BGR2RGB)
res1_median = cv2.cvtColor(img3_median, cv2.COLOR_BGR2RGB)

# SNR using skimage
snr3 = skimage.measure.compare_psnr(img, img3_gaussian,None)
snr4 = skimage.measure.compare_psnr(img, img3_median, None)

# plot
plt.figure(figsize=(20, 20))
plt.subplot(131), plt.imshow(img3_true)
plt.title("Clear Image"), plt.xticks([]), plt.yticks([])

plt.subplot(132), plt.imshow(res1_gaussian)
plt.title("7x7 Gaussian"), plt.xticks([]), plt.yticks([])

plt.subplot(133), plt.imshow(res1_median)
plt.title("7x7 Median"), plt.xticks([]), plt.yticks([])

plt.show()

print("Original vs. Gaussian SNR :",snr3)
print("Original vs. Median SNR   :", snr4)
# --- END of Part 1.1: Q. 4,5,6 ---
```

### Section 1, 1.1 - Q6 Answer:
Median filter seems to give the superior result; because a higher SNR tends to be considerably better. Especially with Salt & Pepper. <br>
Signal/Noise, higher would mean high signal and low noise.
