```python
# Student Name: Nayem Alam
# Student ID:   260743549
# Assignment    1

import numpy as np
import cv2
import matplotlib.pyplot as plt

pepper_img = "./peppers.png"
npepper_img = "./gaussian_noise_peppers.png"

# image filtering
img = cv2.imread(pepper_img)
img2 = cv2.imread(npepper_img)

# filter image using 7x7 Gaussian
res1 = cv2.GaussianBlur(img2, (7,7), 2) # the variance??

# box filter image using 7x7 Gaussian
res2 = cv2.blur(img2, (7,7))

# convert from BGR to RGB for displaying
img = cv2.cvtColor(img, cv2.COLOR_BGR2RGB)
img2 = cv2.cvtColor(img2, cv2.COLOR_BGR2RGB)
res1 = cv2.cvtColor(res1, cv2.COLOR_BGR2RGB)
res2 = cv2.cvtColor(res2, cv2.COLOR_BGR2RGB)

# plot

plt.subplot(132), plt.imshow(res1)
plt.title("7x7 Gaussian"), plt.xticks([]), plt.yticks([])

plt.subplot(133), plt.imshow(res2)
plt.title("7x7 Box"), plt.xticks([]), plt.yticks([])
plt.show()

#skimage.measure.compare_psnr(img, res1,None)



```
