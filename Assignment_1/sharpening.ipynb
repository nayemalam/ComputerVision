```python
# Student Name: Nayem Alam
# Student ID:   260743549
# Assignment    1

import numpy as np
import cv2
import matplotlib.pyplot as plt
import skimage
from skimage import data, io, filters
import math

# --- Part 1.2: Q. 1,2 ---
rice_img = "./rice.png"

img = cv2.imread(rice_img)

# filter image using 5x5 Gaussian
img_gaussian = cv2.GaussianBlur(img, (5,5), math.sqrt(2)) #  sqrt(2) is standard dev., 2 is variance

# box filter image using 5x5
img_box = cv2.blur(img,(5,5))

# sharpen images
g_details = cv2.subtract(img, img_gaussian)
b_details = cv2.subtract(img, img_box)

g_sharpened = cv2.add(img, g_details)
b_sharpened = cv2.add(img, b_details)

# since image is black and white, no point converting from BGR2RGB

# plot
plt.figure(figsize=(20, 20))
plt.subplot(131), plt.imshow(img)
plt.title("Original"), plt.xticks([]), plt.yticks([])

plt.subplot(132), plt.imshow(b_sharpened)
plt.title("Sharpened with Box Filter"), plt.xticks([]), plt.yticks([])

plt.subplot(133), plt.imshow(g_sharpened)
plt.title("Sharpened with Gaussian Filter"), plt.xticks([]), plt.yticks([])

plt.show()

# --- END of Part 1.2: Q. 1,2 ---
```
