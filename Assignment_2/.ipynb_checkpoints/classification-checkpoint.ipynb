### 1. Resize the training images to 128 x 128.

```python
import numpy as np
import cv2 
import os # os commands
from matplotlib import pyplot as plt
import os.path
import glob
from sklearn.neighbors import KNeighborsClassifier

data, label = [],[]

# load directory and index files
imgFile = "./dataset/"
data_path = os.path.join(imgFile,'*g')
files = glob.glob(data_path)

# set dimensions AND preserve aspect ratio
width = 128
height = 128
dim = (width, height)

# find all images in directory
for img in files:
    image = cv2.imread(img)
    image = cv2.cvtColor(image, cv2.COLOR_BGR2RGB)
    filename = os.fsdecode(img)
    filepath = data_path + filename
    resized = cv2.resize(cv2.imread(img),dim)
    resized = cv2.cvtColor(resized, cv2.COLOR_BGR2RGB)
    
    if 'test' in filename:
        continue # ignore any image with word 'test'
    else:
        # put resized images onto numpy array
        data.append(resized)
        
        # Use class label 1 for 'Ford' and -1 for 'Volkswagen'
        if 'ford' in filename:
            label.append(1)
        if 'volks' in filename:
            label.append(-1)
            
# print('labels', np.asarray(label))
#     print(img)

data = np.asarray(data)
print('Resized data - only training images:',data.shape)

# # to visually see all resizes
# plt.figure(figsize=(10, 10))
# plt.subplot(121), plt.imshow(image)
# plt.title("Original image"), plt.axis('on')
# plt.subplot(122), plt.imshow(resized)
# plt.title("Resized image (128x128)"), plt.axis('on')
# plt.show()
```

### 2. Compute HoG features of size (32,32,8). Apply blocknorm in 4x4 cell neighborhood. (Deduce values of other required variables.)

```python
def computeHoG(img, cell_size=(4, 4), block_size=(4, 4), nbins=8):

    img = cv2.cvtColor(img, cv2.COLOR_RGB2GRAY)

    # create HoG Object
    hog = cv2.HOGDescriptor(_winSize=(img.shape[1] // cell_size[1] * cell_size[1],
                                      img.shape[0] // cell_size[0] * cell_size[0]),
                            _blockSize=(block_size[1] * cell_size[1],
                                        block_size[0] * cell_size[0]),
                            _blockStride=(cell_size[1], cell_size[0]),
                            _cellSize=(cell_size[1], cell_size[0]),
                            _nbins=nbins)

    n_cells = (img.shape[0] // cell_size[0], img.shape[1] // cell_size[1])

    # Compute HoG features
    hog_feats = hog.compute(img)\
                   .reshape(n_cells[1] - block_size[1] + 1,
                            n_cells[0] - block_size[0] + 1,
                            block_size[0], block_size[1], nbins) \
                   .transpose((1, 0, 2, 3, 4))  # index blocks by rows first

    # hog_feats now contains the gradient amplitudes for each direction,for each cell of its group for each group.
    # Indexing is by rows then columns.

    # computation for BlockNorm
    gradients = np.full((n_cells[0], n_cells[1], nbins), 0, dtype=float)
    cell_count = np.full((n_cells[0], n_cells[1], 1), 0, dtype=int)

    for off_y in range(block_size[0]):
        for off_x in range(block_size[1]):
            gradients[off_y:n_cells[0] - block_size[0] + off_y + 1,
                      off_x:n_cells[1] - block_size[1] + off_x + 1] += \
                hog_feats[:, :, off_y, off_x, :]
            cell_count[off_y:n_cells[0] - block_size[0] + off_y + 1,
                       off_x:n_cells[1] - block_size[1] + off_x + 1] += 1

    # Average gradients
    gradients /= cell_count
    
    # Preview
    plt.figure(figsize = (10,10))
    plt.subplot(121)
    plt.imshow(img, cmap='gray')
    plt.title("Original Image"), plt.xticks([]), plt.yticks([])

    bin = 0  # angle is 360 / nbins * direction
    plt.subplot(122)
    plt.pcolor(gradients[:, :, bin])
    plt.gca().invert_yaxis()
    plt.gca().set_aspect('equal', adjustable='box')
    plt.title("HOG bin = 0"), plt.xticks([]), plt.yticks([])
    plt.colorbar()
    plt.show()
    
    return gradients
```

<h3><center>HoG Features for Training Images (display)</center></h3>

```python
train = []
for hogs in data:
    train.append(computeHoG(hogs))
train = np.asarray(train)
print('HoG features size: ', train.shape) # to verify features of size requested: (32,32,8)
```

### 3. Fit a nearest neighbor classifier with three neighbors. Use KNeighborsClassifier from sklearn library.

```python
clf = KNeighborsClassifier(n_neighbors=3) # k=3
train = train.reshape(10, -1)
clf.fit(train, label) # fitting the model
print(clf)
```

### 3.2 Reasoning Questions
#### 1. Will HoG work if images of logos undergo random rotation? If yes, which property of HoG avails this feature? If not, discuss the sequence of computer vision technique(s) that can be used in order to regain uniform orientation of the logos.
Histogram of Oriented Gradients (HoG) is a feature descriptor. The histograms in HoG are distributed in directions of gradients (oriented gradients). As a result, though HoG can preprocess any image of any size, but the only constraint is that it requires images to be in a certain orientation and a fixed aspect ratio. Like we did with the 128x128. 
As a result, if we make the logos undergo random rotation, an explicit and exhaustive calculation of all these dimensions would need to be performed.

One way to avail this feature, we can use a Rotation-Invariant Fast Feature (RIFF) descriptor, which is based on HoG feature. RIFF consists of 4 annular cells, where the image gradient orientations in each cell are computed using a centered mask [-1,0,1] and then rotated to the proper angle to achieve rotation invariance.<sup>[1](http://hobbit.acin.tuwien.ac.at/publications/2012_07_isvc_hogpso.pdf)</sup>


#### 2. Consider that you are given a large dataset with three types of images: variable sized squares with either or both diagonals drawn (see Figure 3). How many and which directions of the gradient are sufficient in order to correctly classify the images into three categories?

If we are considering Figure 3 (square sized images), then there are only two diagonals possible. This would mean that <u>two</u> directions of the grradient are sufficient in order to correctly classify the images into three categories. <br>
HoG computes gradients normal to the diagonal, thus in terms of the orientation of directions, we would need to compute the gradients normal to both of the diagonals, and then this will allow us to correctly classify the images.
